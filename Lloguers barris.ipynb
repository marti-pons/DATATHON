```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
lloguers = []
for year in range(2014, 2022):
    file = str(year) + '_lloguer_preu_trim.csv'
    lloguer = pd.read_csv(file)
    lloguer = lloguer.loc[lloguer['Lloguer_mitja'] == 'Lloguer mitjà mensual (Euros/mes)']
    lloguer['Any-Trimestre'] = lloguer['Any'].astype(str) + '-T' + lloguer['Trimestre'].astype(str)
    del lloguer['Lloguer_mitja']
    del lloguer['Any']
    del lloguer['Trimestre']
    del lloguer['Codi_Districte']
    del lloguer['Nom_Districte']
    del lloguer['Nom_Barri']
    lloguers.append(lloguer)

lloguers = pd.concat(lloguers)
lloguers
```

```python
barris = []
for codi_barri in range(1, 74):
    barri = lloguers[lloguers["Codi_Barri"] == codi_barri]
    del barri['Codi_Barri']
    barris.append(barri)

barris[0]
```

```python
X = np.array(range(len(barris[0]))).reshape(-1, 1)  # values converts it into a numpy array
Y = barris[0]['Preu'].values.reshape(-1, 1) # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
```

```python
X_aug = np.array(range(len(barris[0]) + 5 * 4 + 2)).reshape(-1, 1) 
Y_pred = linear_regressor.predict(X_aug)  # make predictions
```
